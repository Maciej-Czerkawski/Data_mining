```python
import numpy as np
import matplotlib.pyplot as plt
import math
from sklearn.naive_bayes import GaussianNB
```

```python
#Avarage values for each class
m1 = np.array([-1,-1])
m2 = np.array([2,4])
m3 = np.array([-2,2])

#Covariance matrix
S = np.array([[4,0],[0,4]])

#Size of each class 
n = 30 

#Generating points from a normal distribution 
r1 = np.random.multivariate_normal(m1,S,n)
r2 = np.random.multivariate_normal(m2,S,n)

#Adding a class vector to points 
r12 = np.concatenate((r1,r2))
y = np.hstack((np.zeros(n), np.ones(n)))
plt.scatter(r12[:,0],r12[:,1], c=y, s=40, cmap="rainbow_r");
plt.title("Plot for points of two classes");
```

```python
#Naive Bayes classificator - ready function 
model = GaussianNB()
model.fit(r12,y)

#Setting up coordinates of points along with classes
plt.scatter(r12[:,0],r12[:,1], c=y,cmap="rainbow_r", s = 40);
lim = plt.axis()
xf, yf = np.meshgrid(np.linspace(lim[0], lim[1],100),
                     np.linspace(lim[2], lim[3],100))

#Setting up all points as uniform sequence of numbers
v = np.c_[xf.ravel(), yf.ravel()]

#Calculating the probabilities of points for each class
Z = model.predict_proba(v)
Z = Z[:, 1].reshape(xf.shape)

plt.contour(xf, yf, Z,[0.5]);
plt.title("Plot for ''ready function'' method ");
```

```python
#Calculating probabilities for each classes 
pi1 = n/(n+n)
pi2 = n/(n+n)
#Calculating avarage values in classes 
me1 = np.mean(r1, axis = 0)
me2 = np.mean(r2, axis = 0)
#Calculating covariance matrix 
se1 = np.cov(r1.T)
se2 = np.cov(r2.T)
#Calculating variations 
var1 = np.var(r1, axis = 0)
var2 = np.var(r2, axis = 0)

#Commonality of the above
pi = [pi1,pi2]
mex = [me1[0],me2[0]]
mey = [me1[1],me2[1]]
varx = [var1[0],var2[0]]
vary = [var1[1],var2[1]]
```

$p(x_i = l|k) = \frac{1}{\sqrt{2*\pi * \sigma_k^2}}*e^{-\frac{(v-\sigma_k^2)}{(2\sigma_k^2)}}$

$\stackrel{\text{^}}{c} = argmax(\pi_k p(x|k) p(y|k)) $

```python
#Naive Bayes classificator according to the theoretical model presented above
def NB(v):
    pxf, pyf = [0,0],[0,0]
    c = np.zeros((2,len(v)))
    for k in range(2): 
        pxf[k] = (1 / np.sqrt(2 * np.pi * varx[k])) * np.exp(-(v[:,0]-mex[k])**2/(2*varx[k]))
        pyf[k] = (1 / np.sqrt(2 * np.pi * vary[k])) * np.exp(-(v[:,1]-mey[k])**2/(2*vary[k]))
        c[k] = pi[k] * pxf[k] * pyf[k]
    return np.array([np.argmax([c[0,n],c[1,n]]) for n in range(len(v))])
```

```python
#Implementation of NB function
y1 = NB(v)
y1 = y1.reshape(xf.shape)
plt.scatter(r12[:,0],r12[:,1], c=y, s=40, cmap="rainbow_r");
plt.contour(xf, yf, y1,[0.5]);
plt.title("Plot with discriminatory line from NB function ");
```

```python
#Comparision of NB() and GaussianNB() on the plot
plt.scatter(r12[:,0],r12[:,1], c=y, s=40, cmap="rainbow_r");
plt.contour(xf, yf, Z,[0.5],colors = 'yellow');
plt.contour(xf, yf, y1,[0.5],colors = 'green');
plt.title("NB() vs GaussianNB()");
ax = plt.gca()
ax.set_facecolor('m')
```

As we can see in the plot above both methods are equivalent. Accuracy of NB() function depends on binning. For this plot set xf and yf as 100 in order to show mutual covering. Bigger values of xf and yf gives same results for both functions.
