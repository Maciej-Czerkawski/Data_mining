```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn import datasets
from itertools import combinations
from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
#Loading data set animals 
df = pd.read_csv('animals.csv')
```

```python
#List of animals 
animals = df.pop('Unnamed: 0')

#Raplacing all NaN with 0
df.fillna(0,inplace=True)

#Adding names to columns 
columns = ("warm-blooded","can fly","vertebrate","endangered","live in groups","have hair")
df.columns = columns
```

```python
#Heatmap for animals data
sns.clustermap(df,cmap="coolwarm",yticklabels=animals,figsize=(8,8),row_cluster=True);
```

Red color means that observation has particular feature. Grey color means that animal doesnt have this feature. Blue is NaN value. As we can see clusterizing process worked well - for example bee and fly in the same category. 

```python
#Loading iris data
iris = pd.read_csv('iris.csv')
```

```python
#List of all possible combinations of [1,2,3,4]
combs = []
for i in range(3):
    combs.append(list(combinations([0,1,2,3], i+2)))
    
combs    
```

```python
#Function to set numerical values for species 
def converter(species):
    if species =='Setosa':
        return 0
    elif species == 'Versicolor':
        return 1
    elif species == 'Virginica':
        return 2
```

```python
#Usage of above function
iris['class'] = iris['variety'].apply(converter)
```

```python
#Accuracy of clasterization for each of combinations(combs)
acc = []
for i in combs:
    for j in range(len(i)):
        x = iris.iloc[:,list((i[j]))].values
        kmeans = KMeans(n_clusters=3)
        kmeans.fit(x)
        acc.append(accuracy_score(iris['class'],kmeans.labels_))
```

```python
#Ploting accuracy depending on combination of variables

combs_lables = ['(1,2)','(1,3)','(1,4)','(2,3)','(2,4)','(3,4)','(1,2,3)'
                ,'(1,2,4)','(1,3,4)','(2,3,4)','(1,2,3,4)']

fig, ax = plt.subplots(figsize=(12,6))
ax.set_xlabel("Combinations of variables")
ax.set_ylabel("Accuracy")
ax.set_title("Accuracy of clustering depending on the combination of variables")
ax.plot(combs_lables, acc, marker='o')
plt.show()
```

We can observe that accuracy of clusterizing fluctuates and some combinations do not allow method to group observations. Major problem for clusterizing based on some combinations of variables probably comes from the fact that at the beginning of clusterization we have wrong first classification that decreases accuracy to under 0.2 level.
